```python
import numpy as np #memasukkan modul
from tabulate import tabulate
```

pada cell di atas, modul numpy dimasukkan agar program di bawah dapat dijalankan

```python
def Jacobi(matrix,initial,tol): #mendefinisikan fungsi Jacobi
    n = np.shape(matrix)[0] #ukuran matrix/banyak variabel
    x0 = initial #mendefinisikan x0 sebagai tebakan awal
    x = x0.copy() #mendefinisikan x = x0 
    
    for i in range(n):
        b = matrix[i,n] #konstanta pada SPL
        for j in range(n): #metode Jacobi
            if j!=i:
                b-=matrix[i,j]*x0[j]
        x[i]=b/matrix[i,i] #hasil aproksimasi Jacobi
    
    tab = [] #membuat list kosong
    iterasi = 1
    
    while np.linalg.norm(x-x0,np.inf)>=tol: #pengecekan kriteria penyetop menggunakan error mutlak
        x0 = x #mendefinisikan x0 sebagai x(k-1)
        x = x0.copy() #mendefinisikan x sebagai x(k), namun awalnya samakan dengan x0
        for i in range(n): #metode Jacobi
            b = matrix[i,n]
            for j in range(n):
                if j!=i:
                    b-=matrix[i,j]*x0[j]
            x[i]=b/matrix[i,i]
        
        tab.append([iterasi, x[0], x[1], x[2]])
        iterasi += 1
            
    tab.append([iterasi, x[0], x[1], x[2]]) #memasukkan nilai yang didapatkan ke dalam array
    print(tabulate(tab, headers = ["iterasi", "x1", "x2", "x3"], tablefmt = "pretty")) #mengubah ke bentuk tabel
    return x #nilai yang direturn ke fungsi ini adalah x(k)

def GaussSeidel(matrix,initial,tol):
    n = np.shape(matrix)[0] #ukuran matrix/banyak variabel
    x0 = initial #definisikan x0 sebagai tebakan awal
    x = x0.copy() #definisikan x = x0
    
    for i in range(n):
        b = matrix[i,n] #konstanta pada SPL
        for j in range(n):
            if j!=i:
                if j<i:
                    b-=matrix[i,j]*x[j]
                else:
                    b-=matrix[i,j]*x0[j]
        x[i]=b/matrix[i,i] #hasil aproksimasi gauss seidel
        
    tab = [] #membuat list kosong
    iterasi = 1
    
    while np.linalg.norm(x-x0,np.inf)>=tol: #pengecekan kriteria penyetop menggunakan error mutlak
        x0 = x #mendefinisikan x0 sebagai x(k-1)
        x = x0.copy() #mendefinisikan x sebagai x(k)
        for i in range(n): #metode gauss seidel
            b = matrix[i,n]
            for j in range(n):
                if j!=i:
                    if j<i:
                        b-=matrix[i,j]*x[j]
                    else:
                        b-=matrix[i,j]*x0[j]
            x[i]=b/matrix[i,i]
        
        tab.append([iterasi, x[0], x[1], x[2]]) 
        iterasi += 1
    
    tab.append([iterasi, x[0], x[1], x[2]]) #memasukkan nilai ke array
    print(tabulate(tab, headers = ["iterasi", "x1", "x2", "x3"], tablefmt = "pretty")) #mengubah ke bentuk tabel
    return x #nilai yang direturn ke fungsi ini adalah x(k)
```
